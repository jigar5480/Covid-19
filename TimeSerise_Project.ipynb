```python
import requests
import pandas as pd
x = requests.get('https://data.covid19india.org/v4/min/timeseries.min.json')
df=pd.DataFrame(x.json()).transpose()
df.head()
df=df.T
```

```python
df1=df.melt(id_vars=None,var_name="State")
df1.head()
```

```python
df2=pd.DataFrame(data=[x for x in df1["value"]],index=df1["State"])
df2.head(1)
```

```python
df2 = df2.reset_index()
```

```python
df2.head(1)
```

```python
df3=df2.melt(id_vars="State",var_name="Dates")
df3
```

```python
df3.isnull().sum()
```

```python
df3.dropna(axis=0,inplace=True)
```

```python
df3.isnull().sum()
```

```python
df3.head()
```

```python
df3["State_Date"]=df3["State"]+"_"+df3["Dates"]
```

```python
df3=df3.drop(columns=["Dates","State"])
```

```python
df3.head()
```

```python
df4=pd.DataFrame(data=[x for x in df3["value"]],index=df3["State_Date"])
df4
```

```python
df4.reset_index(inplace=True)
```

```python
df5=df4.melt(id_vars="State_Date",var_name="Cases")
```

```python
df5.head()
```

```python
df5.isnull().sum()
```

```python
df5.dropna(axis=0,inplace=True)
```

```python
df5.isnull().sum()
```

```python
df5["State_Date_Cases"]=df5["State_Date"]+"_"+df5["Cases"]
```

```python
df5.drop(columns="State_Date",inplace=True)
```

```python
df5.drop(columns="Cases",inplace=True)
```

```python
df5
```

```python
df6=pd.DataFrame(data=[x for x in df5["value"]],index=df5["State_Date_Cases"])
```

```python
df6=df6.reset_index()
```

```python
df6
```

```python
df6.dtypes
```

```python
df6[["State_code","Date","Period"]]=df6["State_Date_Cases"].str.split("_",expand=True)
```

```python
df6.head(2)
```

```python
df6.head()
```

```python
df6.drop(columns="State_Date_Cases")
```

```python
df6.info()
```

```python
df7=df6.pivot_table(
    index=["State_code","Date"],
    columns="Period",
    values=["vaccinated2","vaccinated1","other","recovered","tested","deceased","confirmed"]
)
```
